ðŸ’¡ **Question 1**

Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

</aside>

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def merge_lists(list1, list2):
    new_list_head = None
    current = None

    while list1 and list2:
        if list1.data >= list2.data:
            new_node = Node(list1.data)
            list1 = list1.next
        else:
            new_node = Node(list2.data)
            list2 = list2.next

        if not new_list_head:
            new_list_head = new_node
            current = new_node
        else:
            current.next = new_node
            current = current.next

    # Add remaining nodes if any
    if list1:
        current.next = list1

    if list2:
        current.next = list2

    return new_list_head
```

```python
list1 = Node(1)
list1.next = Node(3)
list1.next.next = Node(5)

list2 = Node(2)
list2.next = Node(4)
list2.next.next = Node(6)

merged_list = merge_lists(list1, list2)

while merged_list:
    print(merged_list.data, end = " ")
    merged_list = merged_list.next
```

ðŸ’¡ **Question 2**

Write a function that takes a list sorted in non-decreasing order and deletes any duplicate nodes from the list. The list should only be traversed once.

For example if the linked list is 11->11->11->21->43->43->60 then removeDuplicates() should convert the list to 11->21->43->60.

</aside>

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def remove_duplicates(head):
    current = head

    while current and current.next:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head

# Create the sorted linked list
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

# Remove duplicates
new_head = remove_duplicates(head)

# Print the modified list
current = new_head
while current:
    print(current.data, end="->")
    current = current.next
```

Given a linked list of sizeÂ **N**. The task is to reverse everyÂ **k**Â nodes (where k is an input to the function) in the linked list. If the number of nodes is not a multiple ofÂ *k*Â then left-out nodes, in the end, should be considered as a group and must beÂ reversed (See Example 2 for clarification).

</aside>

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverse_k_nodes(head, k):
    current = head
    next_node = None
    previous = None
    count = 0

    # Reverse k nodes
    while current and count < k:
        next_node = current.next
        current.next = previous
        previous = current
        current = next_node
        count += 1

    # Recursive call for the remaining nodes
    if next_node is not None:
        head.next = reverse_k_nodes(next_node, k)

    return previous

# Create the linked list
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

# Reverse every 3 nodes
k = 3
new_head = reverse_k_nodes(head, k)

# Print the modified list
current = new_head
while current:
    print(current.data, end="->")
    current = current.next
```

ðŸ’¡ **Question 4**

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

</aside>

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverse_alternate_k_nodes(head, k):
    current = head
    next_node = None
    previous = None
    count = 0

    # Reverse k nodes
    while current and count < k:
        next_node = current.next
        current.next = previous
        previous = current
        current = next_node
        count += 1

    # Connect reversed k nodes with the next group
    if next_node is not None:
        head.next = next_node

    # Skip the next k nodes
    count = 0
    while current and count < k-1:
        current = current.next
        count += 1

    # Recursive call for the next group of k nodes
    if current is not None:
        current.next = reverse_alternate_k_nodes(current.next, k)

    return previous


# Create the linked list
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

# Reverse every alternate 3 nodes
k = 3
new_head = reverse_alternate_k_nodes(head, k)

# Print the modified list
current = new_head
while current:
    print(current.data, end="->")
    current = current.next
```

ðŸ’¡ **Question 5**

Given a linked list and a key to be deleted. Delete last occurrence of key from linked. The list may have duplicates.

</aside>

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def delete_last_occurrence(head, key):
    current = head
    previous = None
    last_occurrence = None
    last_occurrence_previous = None

    # Traverse the linked list to find the last occurrence
    while current:
        if current.data == key:
            last_occurrence = current
            last_occurrence_previous = previous
        previous = current
        current = current.next

    # If the last occurrence is found, delete it
    if last_occurrence:
        if last_occurrence_previous:
            last_occurrence_previous.next = last_occurrence.next
        else:
            head = last_occurrence.next

    return head


# Create the linked list
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(2)
head.next.next.next.next = Node(4)
head.next.next.next.next.next = Node(2)
head.next.next.next.next.next.next = Node(5)

# Delete the last occurrence of key 2
key = 2
new_head = delete_last_occurrence(head, key)

# Print the modified list
current = new_head
while current:
    print(current.data, end="->")
    current = current.next
```

ðŸ’¡ **Question 6**

Given two sorted linked lists consisting ofÂ **N**Â andÂ **M**Â nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

</aside>

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def merge_sorted_lists(head1, head2):
    dummy = Node()  # Dummy node
    prev = dummy
    current1 = head1
    current2 = head2

    # Traverse both lists simultaneously
    while current1 and current2:
        if current1.data <= current2.data:
            prev.next = current1
            current1 = current1.next
        else:
            prev.next = current2
            current2 = current2.next
        prev = prev.next

    # Attach the remaining nodes of the first list, if any
    if current1:
        prev.next = current1

    # Attach the remaining nodes of the second list, if any
    if current2:
        prev.next = current2

    return dummy.next  # Return the head of the merged list

# Create the first sorted linked list
head1 = Node(1)
head1.next = Node(3)
head1.next.next = Node(5)

# Create the second sorted linked list
head2 = Node(2)
head2.next = Node(4)
head2.next.next = Node(6)

# Merge the two lists
merged_head = merge_sorted_lists(head1, head2)

# Print the merged list
current = merged_head
while current:
    print(current.data, end="->")
    current = current.next
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

def reverse_doubly_linked_list(head):
    current = head
    previous = None

    # Reverse the links
    while current:
        next_node = current.next
        current.next = previous
        current.prev = next_node
        previous = current
        current = next_node

    # Update the head pointer
    head = previous

    return head

# Create the doubly linked list
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next

# Reverse the doubly linked list
new_head = reverse_doubly_linked_list(head)

# Print the reversed list (forward direction)
current = new_head
while current:
    print(current.data, end="->")
    current = current.next

# Print the reversed list (backward direction)
current = new_head
while current.next:
    current = current.next
while current:
    print(current.data, end="->")
    current = current.prev
```

Given a doubly linked list and a position. The task is to delete a node from given position inÂ a doubly linked list.

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

def delete_node(head, position):
    if not head:  # Empty list
        return head

    if position == 0:  # Delete the first node
        head = head.next
        if head:
            head.prev = None
        return head

    current = head
    count = 0

    # Traverse to the node at the given position
    while current and count < position:
        current = current.next
        count += 1

    # If the position is beyond the length of the list
    if not current:
        return head

    # Update the neighboring nodes' pointers
    if current.prev:
        current.prev.next = current.next
    if current.next:
        current.next.prev = current.prev

    return head

# Create the doubly linked list
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next

# Delete the node at position 2 (value: 3)
position = 2
new_head = delete_node(head, position)

# Print the modified list (forward direction)
current = new_head
while current:
    print(current.data, end="->")
    current = current.next

# Print the modified list (backward direction)
current = new_head
while current.next:
    current = current.next
while current:
    print(current.data, end="->")
    current = current.prev
```
